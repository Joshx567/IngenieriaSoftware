```python
import numpy as np 
import matplotlib.pyplot as pl
```

```python
data = np.loadtxt('datos01.txt', delimiter = ",")
data
```

```python
data.shape
```

```python
x = data[:, 0]
y = data[:, 1]
pl.scatter(x,y)
pl.grid()
```

```python
def get_j(t,x,y):
    m = x.shape[0]
    h = t[0] + t[1]*x
    error = h - y
    j = 1/(2*m)*error**2
    return j.sum()
```

```python
get_j(np.array([0, 0]),x,y)
```

```python
def get_gradiente(t,x,y):
    m = x.shape[0]
    h = t[0] + t[1]*x
    error = h - y
    grad0 = 1/m*error
    grad1 = 1/m* error*x
    return np.array([(grad0.sum()),grad1.sum()])
```

```python
get_gradiente(np.array([0, 0]),x,y)
```

```python
fig = pl.figure()
ax = fig.add_subplot(111, projection='3d')

# Generamos los valores de T0 y T1
T0 = np.linspace(-10, 10, 100)
T1 = np.linspace(-1, 4, 100)
T0, T1 = np.meshgrid(T0, T1)

# Inicializamos J
J = np.zeros((100, 100))

# Rellenamos la matriz J
for i in range(100):
    for j in range(100):
        t = np.array([T0[i, j], T1[i, j]])  # Corregido el formato del array
        J[i, j] = get_j(t, x, y)  # Asegrate que X y Y estn definidas

# Agregamos un punto con 'scatter'
ax.scatter(0, 0, get_j(np.zeros(2), x, y), color='red')

# Creamos la grfica de superficie
ax.plot_surface(T0, T1, J, cmap='viridis')

# Mostramos la grfica
pl.show()
```
